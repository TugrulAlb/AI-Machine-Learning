```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('advertising.csv')
```

```python
df.head()
```

```python
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']             
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model_tv = LinearRegression().fit(X_train[['TV']], y_train)
model_radio = LinearRegression().fit(X_train[['Radio']], y_train)
model_newspaper = LinearRegression().fit(X_train[['Newspaper']], y_train)
```

```python
(model_tv.coef_[0], model_radio.coef_[0], model_newspaper.coef_[0])
```

```python
y_pred_tv = model_tv.predict(X_test[['TV']])
y_pred_radio = model_radio.predict(X_test[['Radio']])
y_pred_newspaper = model_newspaper.predict(X_test[['Newspaper']])
```

```python
plt.figure(figsize=(12, 7))
plt.plot(y_test.values, label='Gerek Satlar', marker='o')
plt.plot(y_pred_tv, label='TV Tahmini', marker='x')
plt.plot(y_pred_radio, label='Radyo Tahmini', marker='^')
plt.plot(y_pred_newspaper, label='Gazete Tahmini', marker='s');
```
